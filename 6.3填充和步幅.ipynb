```python
import torch 
from torch import nn
```

```python
#方便起见，我们定义一个计算卷积的函数
#此函数初始化卷积层权重，并堆输入和输出扩大和缩减相应的维度
def comp_conv2d(conv2d,X):
    #这里的(1,1)表示批量大小和通道数都是1
    X = X.reshape((1,1) + X.shape)
    #X.shape 的本质对于 PyTorch 张量 X，X.shape 返回的是一个 元组，用于表示张量的维度。
    #例如，当 X 是一个 8x8 的张量时，X.shape 的值是 (8, 8)（一个包含两个元素的元组）。
    #元组的 + 运算符用于 拼接两个元组，生成一个新的元组。
    #a = (1, 1)
    # b = (8, 8)
    # c7 = a + b  # 结果是 (1, 1, 8, 8)
    
    Y = conv2d(X)
    #省略前两个维度：批量大小和通道
    return  Y.reshape(Y.shape[2:])

#注意，这里每侧边都填充了1行或1列，因此总共添加了2行或2列
conv2d = nn.Conv2d(1,1,kernel_size = 3,padding = 1)
#kernel_size=3：卷积核大小为3x3。
#padding=1：在输入特征图的上下左右各填充 1 行 / 列（用于保持输出尺寸）

X = torch.rand(size = (8,8))
#torch.rand(size=(8, 8))生成一个8x8的随机张量，模拟单通道的输入特征图（如8x8的灰度图像）

comp_conv2d(conv2d,X).shape
```

```python
conv2d = nn.Conv2d(1,1,kernel_size=(5,3),padding = (2,1))
#高度和宽度两侧的填充分为2和1
comp_conv2d(conv2d,X).shape
```

```python
conv2d = nn.Conv2d(1,1,kernel_size=3,padding=1,stride=2)
comp_conv2d(conv2d,X).shape
```

```python
conv2d = nn.Conv2d(1,1,kernel_size=(3,5),padding=(0,1),stride=(3,4))
comp_conv2d(conv2d,X).shape
```
